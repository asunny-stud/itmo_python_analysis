```python
import pandas as pd
import numpy as np


df = pd.read_csv("./beverage_r.csv", sep=";", index_col='numb.obs')
```

```python
df.head()
```

```python
len(df)
```

```python
np.random.seed(seed=42)
r1 = np.random.uniform(low = 0.0, high=1.0, size=len(df))
r1
```

## Задача 1
Разделить таблицу данных так, чтобы в обучающую выборку попало 66% наблюдений, а в тестовую выборку 33% наблюдений



```python
df_train = df.iloc[r1 > np.quantile(r1, 1/3), :]
df_test = df.iloc[np.logical_not(r1 > np.quantile(r1, 1/3)), :]
print(f'Размер обучающей выборки: {len(df_train)}. Процент наблюдений: {np.round(len(df_train)/len(df), 4)}')
print(f'Размер тестовой выборки: {len(df_test)}. Процент наблюдений: {np.round(len(df_test)/len(df), 4)}')
```

## Задача 2
Разделить таблицу данных на четыре одинаковых части.

```python
df_1 = df.iloc[r1 < np.quantile(r1, 0.25), :]
df_2 = df.iloc[(r1 >= np.quantile(r1, 0.25)) & (r1 < np.quantile(r1, 0.5)), :]
df_3 = df.iloc[(r1 >= np.quantile(r1, 0.5)) & (r1 < np.quantile(r1, 0.75)), :]
df_4 = df.iloc[r1 >= np.quantile(r1, 0.75), :]

print(f'Выборка 1: {len(df_1)}, Процент наблюдений: {np.round(len(df_1)/len(df), 4)}')
print(f'Выборка 2: {len(df_2)}, Процент наблюдений: {np.round(len(df_2)/len(df), 4)}')
print(f'Выборка 3: {len(df_3)}, Процент наблюдений: {np.round(len(df_3)/len(df), 4)}')
print(f'Выборка 4: {len(df_4)}, Процент наблюдений: {np.round(len(df_4)/len(df), 4)}')
```

# Задача 3
Разделить таблицу данных на две стратифицированные подвыборки.
По переменным кока-кола и минералка.

```python
df['strat'] = 0
df.loc[df['COKE'] == 1, 'strat'] = 1
df.loc[df['TAB'] == 1, 'strat'] = 2
df.head()
```

```python
df[['COKE', 'TAB']].value_counts(normalize = True)
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(
    df,
    test_size=0.5,
    stratify=df['strat'],
    random_state=42
)
```

```python
train_df = train_df.drop('strat', axis=1)
train_df[['COKE', 'TAB']].value_counts(normalize = True)
```

```python
test_df = test_df.drop('strat', axis=1)
test_df[['COKE', 'TAB']].value_counts(normalize = True)
```
