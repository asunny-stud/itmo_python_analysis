## Задача 1

**2.14.** Из колоды карт (52 листа) наудачу вынимаются три карты.
 Найти вероятность того, что:
1. среди них окажется ровно один туз;
2. среди них окажется хотя бы один туз;
3. это будут тройка, семерка и туз (в любом порядке).



Общее число исходов = $C_{52}^3 = \frac{n!}{k!(n-k)!} = \frac{52\cdot51\cdot50}{3\cdot2}= 22100$

1. $P_{ровно \space 1 \space туз} = \frac{C_4^1 ⋅C_{48}^2}{C_{52}^3} = 4⋅\frac{47⋅48}{2}⋅\frac{1}{22100}=\frac{1128}{5525}\approx 0.204163$



2. $P_{хоть\space 1 \space туз} = 1-P_{нет \space тузов} = 1 - \frac{C_{48}^3}{C_{52}^3}=1-\frac{48\cdot47⋅48}{6}⋅\frac{1}{22100}=1-\frac{4324}{5525}=\frac{1201}{5525}\approx0.217376$


3. $P_{3,7,A}=\frac{C_4^1 ⋅C_4^1 ⋅C_4^1}{C_{52}^3} = \frac{4⋅4⋅4}{22100} = \frac{16}{5525}\approx0.002896$

```python
import pandas as pd
import numpy as np
import random

def aces (cards):
  aces = [13, 26, 39, 52]
  return len((set(cards) & set(aces)))

def threes (cards):
  threes = [2, 15, 28, 41]
  return len((set(cards) & set(threes)))

def sevens (cards):
  sevens = [6, 19, 32, 45]
  return len((set(cards) & set(sevens)))

deck = [i for i in range(1, 53)]

random.seed(a=42, version=2)

n_exp = 1000000

p1 = 0
p2 = 0
p3 = 0

for i in range(n_exp):
  k = random.sample(deck,3)

  a = aces(k)
  t = threes(k)
  s = sevens(k)

  if a == 1:
    p1 += 1
  if a > 0:
    p2 += 1
  if a == 1 and t == 1 and s == 1:
    p3 += 1

print ('Вероятность вытянуть ровно один туз = ', np.round(p1/n_exp, 6))
print ('Вероятность вытянуть хотя бы один туз = ', np.round(p2/n_exp, 6))
print ('Вероятность вытянуть тройку, семерку и туз (в любом порядке) = ', np.round(p3/n_exp, 6))
```

## Задача 2

**2.48.** Бросаются 5 игральных костей. Найти вероятность того, что
 по меньшей мере на
1. двух;
2. трех

из них выпадут одинаковые грани.


Общее число исходов $ = 6^5 = 7776$

1. Вероятность, что хотя бы на 2х костях выпадут одинаковые грани.

    $\overline{A} =$ на всех костях разные грани.

    $P_1 = 1 - \overline{A} = 1 - \frac{6⋅5⋅4⋅3⋅2}{7776} = 1 - \frac{5}{54} = \frac{49}{54}  ≈ 0.907407$

2. Вероятность, что хотя бы на 3х костях выпадут одинаковые грани.
    Возможно 3 случая:
    - Все 5 кубиков имеют одинаковые грани:
        $N_5 = 6$
    - 4 кубика имеют одинаковые грани: $N_4 = 6\cdot C_5^4 \cdot 5 = 6 ⋅ 5 ⋅ 5 = 150$

    - 3 кубика имеют одинаковые грани: $N_3 = 6 ⋅ C_5^3 ⋅ 5 ⋅ 5 = 6 ⋅ 10 ⋅ 25 = 1500$



    $P_2 = \frac{5+150+1500}{7776} = \frac{23}{108} ≈ 0.212963$

```python
from collections import Counter
def similar_count(exp):
  counter = Counter(exp)
  return max(counter.values() if counter else 0)

exp = list(range(1,7))

random.seed(a=42, version=2)

n_exp = 1000000

p1 = 0
p2 = 0

for i in range(n_exp):
  a = random.choices(exp, k=5)
  c = similar_count(a)

  if c >= 2:
    p1 += 1
  if c >= 3:
    p2 += 1

print('Вероятность, что не менее 2 одинаковых граней', np.round(p1/n_exp, 6))
print('Вероятность, что не менее 3 одинаковых граней', np.round(p2/n_exp, 6))
```
