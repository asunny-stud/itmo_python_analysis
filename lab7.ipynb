```python
import numpy as np
import pandas as pd


from scipy.stats import norm, chi2_contingency
import statsmodels.api as sm
```

### Таблица 1. Результаты Надаля на глиняных и других кортах


Нулевая гипотеза: Покрытие корта не влияет на результат игры, разница в проценте побед случайна.

```python
s1 = 1660           #  число успехов     выборка А
n1 = 2523      #  число испытаний   выборка А
s2 = 3658             #  число успехов     выборка Б
n2 = 6373     #  число испытаний   выборка Б
```

Используем  proportions_ztest


```python
z1, p_value1 = sm.stats.proportions_ztest([s1, s2], [n1, n2])
```

z-метка и p-значение

```python
print(['{:.12f}'.format(b) for b in (z1, p_value1)])
```

Z-статистика показывает, на сколько стандартных отклонений наша наблюдаемая разница в винрейте (65.8% против 57.4%) отличается от нуля (то есть от ситуации, когда разницы нет). Чем больше число, тем сильнее разница.

Значимость меньше 0.05, поэтому мы отвергаем нулевую гипотезу.

Разница в проценте побед Надаля на грунтовых и других кортах статистически значима.

То есть, Надаль на грунте играет значительно лучше (эффективнее), чем на других покрытиях.

### Таблица 2. Результаты Надаля на глиняных и других кортах — геймы, в которых он подавал


Нулевая гипотеза: Покрытие корта не влияет на вероятность Надаля выиграть гейм, когда он подает.

```python
s1 = 1090
n1 = 1261
s2 = 2747
n2 = 3193
```

```python
z1, p_value1 = sm.stats.proportions_ztest([s1, s2], [n1, n2])
print(['{:.12f}'.format(b) for b in (z1, p_value1)])
```

Z-значение - очень маленькое число. Разница в проценте выигранных геймов на своей подаче между грунтом и другими покрытиями очень мала по сравнению с естественным разбросом данных.

Значимость больше 0.05, поэтому мы принимаем нулевую гипотезу.
Несмотря на то, что Надаль чаще выигрывает матчи на грунте, анализ его геймов на подаче показывает, что покрытие корта не оказывает значимого влияния на его способность брать свою подачу. Процент выигранных подач примерно одинаков на любом покрытии.
